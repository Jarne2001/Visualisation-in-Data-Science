```python
import pandas as pd
import os

# Path to your data directory
base_path = '/content/drive/My Drive/Visualisation in data science/data/'

# Initialize an empty list to store all dataframes
all_dfs = []

# Loop through years 2001 to 2018
for year in range(2001, 2019):
    # Construct the file path
    file_path = f"{base_path}madrid_{year}.csv"

    try:
        # Read the CSV file
        df = pd.read_csv(file_path)

        # Add a year column
        df['year'] = year

        # Append to our list
        all_dfs.append(df)

        print(f"Successfully loaded data for {year}, shape: {df.shape}")

    except FileNotFoundError:
        print(f"Warning: File for year {year} not found at {file_path}")
    except Exception as e:
        print(f"Error processing file for year {year}: {e}")

# Concatenate all dataframes
if all_dfs:
    combined_df = pd.concat(all_dfs, ignore_index=True)

    # Output path
    output_path = f"{base_path}madrid_combined.csv"

    # Export as CSV
    combined_df.to_csv(output_path, index=False)

    print(f"\nCombined data successfully exported to {output_path}")
    print(f"Final dataset shape: {combined_df.shape}")
    print("\nFirst few rows of the combined dataset:")
    print(combined_df.head())
else:
    print("No data was loaded. Please check the file paths.")
```
